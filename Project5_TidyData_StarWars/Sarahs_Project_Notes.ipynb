```python
import pandas as pd
import altair as alt
import numpy as np 

url = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/star-wars-survey/StarWars.csv'
sw_all = pd.read_csv(url, encoding='ISO-8859-1')
sw_responses = pd.read_csv(url, encoding='ISO-8859-1', header=None, skiprows=2) # skips first 2 rows
sw_questions = pd.read_csv(url, encoding='ISO-8859-1', header=None, nrows=2) # reads in first 2 rows
```

```python
questions = (sw_questions.iloc[0,:]
                .replace('Have you seen any of the 6 films in the Star Wars franchise?', 'have_seen_any')
                .replace('Do you consider yourself to be a fan of the Star Wars film franchise?', 'fan_sw')
                .replace('Which of the following Star Wars films have you seen? Please select all that apply.', 'seen_any_')
                .replace('Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.', 'rank_preference_')
                .replace('Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.', 'characters_')
                .replace('Which character shot first?', 'shot_first')
                .replace('Are you familiar with the Expanded Universe?', 'familiar')
                .replace('Do you consider yourself to be a fan of the Expanded Universe?ÂŒÃ¦', 'expanded_universe')
                .replace('Do you consider yourself to be a fan of the Star Trek franchise?', 'star_trek_franchise')
                .replace('Location (Census Region)', 'location')
                .str.lower()
                .str.replace(' ', '_')
                .ffill()
            )

response_questions = (sw_questions.iloc[1,:]
                        .replace('Response', '')
                        .str.replace('Star Wars: Episode ', '')
                        .str.lower()
                        .str.replace(' ', '_')
                        .fillna('')
                    )

column_names = questions + response_questions

column_names
```

```python
# Create a new column that converts age range to one number
#q3.age.str.split('-', expand = True)
# Using minimum numbers in algorithm
ml_age = (q3.age
    .str.split('-', expand = True)
    # rename columns
    .rename(columns = {0:'age_min', 1:'age_max'})
    # get rid of the greater than symbol in front of the 60
    .apply(lambda x: x.str.replace('> ', ''))
    # convert string to float
    .astype('float')
    # extract just the minimum column
    .age_min
    )
ml_age
```

```python
# Convert each school grouping into a number and convert string to float
ml_school = (q3.education. 
        str.replace('Less than high school degree', '9').
        str.replace('High school degree', '12').
        str.replace('Some college or Associate degree', '14').
        str.replace('Bachelor degree', '16').
        str.replace('Graduate degree', '20').
        astype('float')
        )
ml_school
```

```python
# Create a new colum that converts income ranges into a single number
print(q3.household_income.value_counts())

ml_income = (q3.household_income
    .str.split('-', expand = True)
    # rename columns
    .rename(columns = {0:'income_min', 1:'income_max'})
    # get rid of the $ in front of each number
    .apply(lambda x: x.str.replace('$', ''))
    # get rid of + after $150,000
    .apply(lambda x: x.str.replace('+', ''))
    # get rid of comma in each number
    .apply(lambda x: x.str.replace(',', ''))
    # convert string to float
    .astype('float')
    # extract just the minimum column
    .income_min
    )
ml_income
```
