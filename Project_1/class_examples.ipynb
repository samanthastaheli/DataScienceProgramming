## Create Altair Chart
### Requies:
- alt.Chart()
- .mark_
    - mark_circle and mark_line start at 0
    - mark_line starts at value data starts at
- .encode()
### Extra Stuff To Make More Clear (The So What?):
- title using .properties()
- remove commas for year axis labels
    - use alt.X() or alt.Y()
    - then use alt.Axis(format = 'put format here')
    - format = 'd' makes decimal
    - format = '%' makes percent
- change axis titles




```python
bob_chart = (alt.Chart(bob)
    .mark_line()
    .encode(x = alt.X('year', axis = alt.Axis(format = 'd', title = 'Year')), y = 'Total')
    .properties(
        height = 150,
        width = 500,
        title = {'text': 'Chart About Bob', 'subtitle': 'A really cool subtitle for Bob!'}
    )
)
```

```python
birthday = bob.query('year == 1960')

birthday_point = (alt.Chart(birthday)
    .mark_circle(color = 'red', size = 50)
    .encode(x = 'year', y = 'Total')
    )

bob_chart + birthday_point
```

## Chart for my name

```python
my_name_list = ['Sam', 'Samantha', 'Sammy', 'Sami', 'Samuel']

# my_names = (names.query('name in @my_name_list')
#     .groupby('name')
#     .mean()
#     .filter(['WA', 'Total'])
#     .sort_values('WA', ascending=False))

my_names = (names.query('name in @my_name_list'))

my_name_chart = (alt.Chart(my_names)
    .mark_line()
    .encode(
        x = 'year',
        y = 'Total',
        color = 'name')
    .properties(
        height = 150,
        width = 500,
        title = {'text': 'Chart About the Variations Sam'}
    )
        )
my_name_chart
```

## Chart for teamates names

```python
names_list = ['Sarah', 'Samantha', 'Thomas']

team_names = (names.query('name in @names_list'))
```

```python
name_chart = (alt.Chart(team_names)
    .mark_line(interpolate='step-after')
    .encode(
        x = alt.X('year', axis = alt.Axis(format = 'd', title = 'Year')),
        y = 'Total',
        color = 'name')
    .properties(
        height = 150,
        width = 500,
        title = {'text': 'Chart About Samantha', 'subtitle': 'Data for the Name "Samantha"'}
    )
        )
name_chart
```

```python
name_chart = (alt.Chart(team_names)
    .mark_area(opacity=0.3)
    .encode(
        x = alt.X('year', axis = alt.Axis(format = 'd', title = 'Year')),
        y = 'Total',
        color = 'name')
    .properties(
        height = 150,
        width = 500,
        title = {'text': 'Chart About Samantha', 'subtitle': 'Data for the Name "Samantha"'}
    )
        )
name_chart
```
