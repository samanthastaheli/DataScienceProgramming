## Question 1

### byui id: "idbyuid"

```python
byui_players = dw.query('byuidss/cse-250-baseball-database', 
'''
SELECT salaries.playerid, salaries.salary, salaries.teamid, salaries.yearid, collegeplaying.schoolid
FROM salaries AS sal
    JOIN collegeplaying AS cp ON sal.playerid = cp.playerid
 WHERE collegeplaying.schoolid = "idbyuid"
 ORDER BY salaries.salary
''')
byui_players.dataframe
```

## Question 2

### q2a 

```python
q2a = dw.query('byuidss/cse-250-baseball-database', 
'''
SELECT batting.playerid, batting.yearid, (batting.h / batting.ab) AS batting_av 
FROM batting
WHERE batting.ab >= 1
GROUP BY batting.playerid, batting.yearid
ORDER BY batting_av DESC, batting.playerid
LIMIT 5
''')
q2a.dataframe
```

### q2c

## Question 3

### UW Hall of Fame table

```python
uw_table = dw.query('byuidss/cse-250-baseball-database', 
'''
SELECT collegeplaying.schoolid, HallOfFame.playerid, HallOfFame.yearid
FROM HallOfFame AS hof
    JOIN collegeplaying AS cp ON hof.playerid = cp.playerid
 WHERE collegeplaying.schoolid = "washington" AND collegeplaying.schoolid = "washst"
 GROUP BY collegeplaying.schoolid
 ORDER BY HallOfFame.playerid
''')
uw_table.dataframe
```
